Importing Necessary Libraries

```python
# Step 1: Import necessary libraries
import nltk
from nltk.corpus import movie_reviews
import random
import pandas as pd
from nltk.classify import NaiveBayesClassifier
from nltk.classify.util import accuracy
from IPython.display import display # Import display
```

 Download required NLTK datasets

```python
nltk.download('movie_reviews')
nltk.download('punkt')
nltk.download('punkt_tab')
```

Viewing the corpus (movie reviews)

```python
# Build list of (fileid, category, review_text)
data = []
for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        words = " ".join(movie_reviews.words(fileid))  # combine words into full review
        data.append((fileid, category, words))

# Convert into DataFrame
df = pd.DataFrame(data, columns=["fileid", "category", "review_text"])

# Show first few rows
print(df.head())
```

Step 2: Prepare the dataset

```python

documents = [(list(movie_reviews.words(fileid)), category)
             for category in movie_reviews.categories() #category iteration is the outer loop ,iterates over the 2 categories of the corpus
             for fileid in movie_reviews.fileids(category)] #here fileid is the inner loop,iterates over the fileids against that specific category taken in the outer loop
#Suppose 1st itr: Category taken:'neg' , then fileid taken is the id against that 'neg' category,similarly for 'pos'
random.shuffle(documents)
df=pd.DataFrame(documents)
print(display(df))
```

Step 3: Feature extraction function

```python
#Extract the word features
def extract_features(words):
  return {word.lower(): True for word in words}
```

Step 4: Create feature sets

```python
feature_sets = [(extract_features(doc), category) for (doc, category) in documents]
print(display(pd.DataFrame(feature_sets)))
```

Step 5: Split into training and testing data (80% training, 20% testing)

```python
train_size = int(len(feature_sets) * 0.8)
train_set, test_set = feature_sets[:train_size], feature_sets[train_size:]
```

Step 6: Train the Naive Bayes Classifier

```python
classifier = NaiveBayesClassifier.train(train_set)
```

 Step 7: Check accuracy

```python
print("Accuracy:", accuracy(classifier, test_set))
```

Step 8: Interactive sentiment detection

```python
while True:
    user_input = input("Enter your text (type 'quit' to exit): ")
    if user_input.lower() == 'quit':
        break
    words = nltk.word_tokenize(user_input)
    features = extract_features(words)
    sentiment = classifier.classify(features)
    print("Detected sentiment:", sentiment)
```
